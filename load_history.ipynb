```python
import numpy as np

load_dict = np.load('./20231207_dpp_history1.npy',allow_pickle=True).item()
```

```python
from sklearn.datasets import make_blobs  

np.random.seed(0)
num_samples = 1000
num_features = 2
num_clusters = 4

data, labels = make_blobs(n_samples=num_samples, n_features=num_features, centers=num_clusters, random_state=0)
```

```python
final_classification_1 = load_dict['clus_alloc'][1]
final_classification_1 = np.concatenate(final_classification_1)
final_classification_2 = load_dict['clus_alloc'][20]
final_classification_2 = np.concatenate(final_classification_2)
final_classification_3 = load_dict['clus_alloc'][49]
final_classification_3 = np.concatenate(final_classification_3)
```

```python
from collections import Counter

def count_unique(arr):
    unique_elements, counts = np.unique(arr, return_counts=True)
    element_counts = Counter(dict(zip(unique_elements, counts)))
    for element, count in element_counts.items():
        print(f"元素 {element} 出现的次数是: {count}")
count_unique(final_classification_1)
count_unique(final_classification_2)
count_unique(final_classification_3)
```
